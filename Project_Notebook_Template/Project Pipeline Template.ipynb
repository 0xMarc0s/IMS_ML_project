```python
import pandas as pd

data_raw = pd.read_csv('../data/obesity_train.csv')
```

```python
data_raw.head()
```

```python
data_filtered = data_raw.drop(columns=["marrital_status"], axis=1)
data_filtered = data_filtered.dropna(how='all').loc[data_raw.isna().any(axis=1)]
data_filtered
```

```python
data_raw.info()
```

```python
import pandas as pd

def classify_bmi_comprehensive(row):
    """
    Classify BMI based on age and BMI value.
    
    Input:
    row: A Pandas row with 'weight', 'height', and 'age' columns.

    Output:
    Returns a string that classifies the individual into BMI categories.
    """
    try:
        # Check if weight and height are valid
        if row['height'] <= 0 or row['weight'] <= 0:
            return 'Invalid data'
        
        # Calculate BMI
        bmi = row['weight'] / (row['height'] ** 2)
        
        # Classify based on age group
        if row['age'] < 2 or row['age'] > 120:  # Handle unrealistic ages
            return 'Invalid age'
        
        # Age group: Children (2-19 years)
        if 2 <= row['age'] < 20:
            if bmi < 14:
                return 'Underweight'
            elif 14 <= bmi < 18:
                return 'Normal weight'
            elif 18 <= bmi < 21:
                return 'Overweight'
            else:
                return 'Obese'
        
        # Age group: Adults (20-64 years)
        elif 20 <= row['age'] < 65:
            if bmi < 18.5:
                return 'Underweight'
            elif 18.5 <= bmi < 24.9:
                return 'Normal weight'
            elif 25 <= bmi < 29.9:
                return 'Overweight'
            else:
                return 'Obese'

    except Exception as e:
        return f"Error: {e}"  # Catch and report any errors (e.g., missing data)
```

```python
data_filtered = data_filtered.dropna()
df_with_dummies = data_filtered.drop(columns=["obese_level"], axis=1)
df_with_dummies['bmi_class'] = df_with_dummies.apply(lambda row: classify_bmi_comprehensive(row), axis=1)
df_with_dummies = pd.get_dummies(df_with_dummies, drop_first=True).astype('float32')
df_with_dummies['obese_level'] = data_filtered[data_filtered['id'] == df_with_dummies['id']].obese_level
df_with_dummies.drop(columns=["id"], inplace=True)
df_with_dummies
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X = df_with_dummies.drop(columns=["obese_level"], axis=1)
y = df_with_dummies.obese_level
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42,stratify=y)


from sklearn.ensemble import RandomForestClassifier as RandomForest
classifier = DecisionTreeClassifier(criterion='entropy', max_depth=4)

from sklearn.metrics import f1_score

classifier.fit(X_train, y_train)
print(classifier.score(X_test, y_test))
print(classifier.score(X_train, y_train))

print(f1_score(y_test, classifier.predict(X_test), average='weighted'))
```

```python
feature_importances = pd.DataFrame({
    'Feature': classifier.feature_names_in_,
    'Importance': classifier.feature_importances_
})

# Sort the feature importances from highest to lowest
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

print(feature_importances)
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, classifier.predict(X_test))
print(report)
```
